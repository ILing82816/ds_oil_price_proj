```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# path
path_prefix = 'D:/USA 2020 summer/Machine Learning/ds_oil_price_proj'
```

```python
from pandas.plotting import register_matplotlib_converters
# Plot
fig, axes = plt.subplots(nrows=19, ncols=1, dpi=120, figsize=(6,25))
for i, ax in enumerate(axes.flatten()):
 data = df[df.columns[i+1]]
 ax.plot(data, color='red', linewidth=1)
# Decorations
 ax.set_title(df.columns[i+1])
 ax.xaxis.set_ticks_position('none')
 ax.yaxis.set_ticks_position('none')
 ax.spines['top'].set_alpha(0)
 ax.tick_params(labelsize=6)
plt.tight_layout();
```

```python
# check wti price autocorrelation: ACF/PACF
print("Check ACF/PACF ...")
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, ax = plt.subplots(2, figsize=(10,6))
ax[0] = plot_acf(df['wti_price'], ax=ax[0])
ax[1] = plot_pacf(df['wti_price'], lags=20, ax=ax[1])
plt.savefig(os.path.join(path_prefix, 'Figure/ACF_PACF.png'))
plt.show()
##Based ACF and PACF, there is AR(3). 
```
